```python
import joblib
import numpy as np
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Load Model SVM & RF
svm_model = joblib.load("/content/drive/MyDrive/ML dicoding/pengembangan ml/NLP/analisis sentimen/submission/File model/svm_model.pkl")
rf_model = joblib.load("/content/drive/MyDrive/ML dicoding/pengembangan ml/NLP/analisis sentimen/submission/File model/rf_model.pkl")

# Load TF-IDF Vectorizer (untuk SVM & RF)
tfidf_vectorizer = joblib.load("/content/drive/MyDrive/ML dicoding/pengembangan ml/NLP/analisis sentimen/submission/File model/tfidf_vectorizer.pkl")

# Load Model LSTM & Tokenizer
lstm_model = load_model("/content/drive/MyDrive/ML dicoding/pengembangan ml/NLP/analisis sentimen/submission/File model/lstm_model.h5")
tokenizer = joblib.load("/content/drive/MyDrive/ML dicoding/pengembangan ml/NLP/analisis sentimen/submission/File model/tokenizer.pkl")

# Parameter yang sama seperti training
max_len = 100  # Panjang maksimal teks untuk LSTM

# Fungsi untuk Prediksi SVM & RF
def predict_ml(text, model):
    text_tfidf = tfidf_vectorizer.transform([text])  # Ubah teks ke vektor TF-IDF
    pred_label = model.predict(text_tfidf)[0]  # Prediksi kelas
    label_mapping = {0: "negative", 1: "neutral", 2: "positive"}
    return label_mapping[pred_label]

# Fungsi untuk Prediksi LSTM
def predict_lstm(text):
    seq = tokenizer.texts_to_sequences([text])
    pad = pad_sequences(seq, maxlen=max_len, padding='post')
    pred = lstm_model.predict(pad)
    pred_label = np.argmax(pred)  # Ambil label dengan probabilitas tertinggi
    label_mapping = {0: "negative", 1: "neutral", 2: "positive"}
    return label_mapping[pred_label]

# Contoh Testing
input_text = "aplikasinya oke, tolong ditingkatkan"

# Prediksi dengan semua model
result_svm = predict_ml(input_text, svm_model)
result_rf = predict_ml(input_text, rf_model)
result_lstm = predict_lstm(input_text)

print(f"Hasil Prediksi SVM: {result_svm}")
print(f"Hasil Prediksi RF: {result_rf}")
print(f"Hasil Prediksi LSTM: {result_lstm}")
```
